**451 градус по Фаренгейту**

Відомий американський письменник Рей Бредбері має роман «451 градус за Фаренгейтом». Напишіть програму, яка визначає, яку температуру за шкалою Цельсія відповідає вказане значення за шкалою Фаренгейта.

Використовуйте формулу для перекладу:

*C = 5/9 * (F - 32)*

**Формат вхідних даних**

На вхід до програми подається дійсне число градусів за шкалою Фаренгейта.

**Формат вихідних даних**

Програма має вивести кількість градусів за шкалою Цельсія.

**Sample Input 1:**

450.55

**Sample Output 1:**

232.5277777777778

**Sample Input 2:**

113.75

**Sample Output 2:**

45.41666666666667


```python
fahrenheit = float(input("Введіть температуру в градусах за шкалою Фаренгейта: "))
celsius = 5/9 * (fahrenheit - 32)
print("Температура в градусах Цельсія:", celsius)
```

**Сортування трьох**

Напишіть програму, яка приймає на вхід три цілих числа, кожне на окремому рядку, та сортує їх в порядку від більшого до меншого.

**Формат вхідних даних**

На вхід до програми подається три цілих числа, кожне на окремому рядку.

**Формат вихідних даних**

Програма має вивести три числа, кожне на окремому рядку, упорядкованих від більшого до меншого.

**Sample Input 1:**

132

129

135

**Sample Output 1:**

135

132

129

**Sample Input 2:**

150

160

156

**Sample Output 2:**

160

156

150

```python
num1, num2, num3 = map(int, input("Введіть 3 числа через пробіл: ").split())

numbers = [num1, num2, num3]
numbers.sort(reverse=True)

print("Відсортовані числа:", *numbers)
```

**Коректний email**

Вважатимемо email адресу коректною, якщо в ньому є символ собачки (@) і точки. Напишіть програму, яка перевіряє коректність email адреси.

**Формат вхідних даних**

На вхід до програми подається один рядок – email адреса.

**Формат вихідних даних**

Програма повинна вивести рядок «YES», якщо адреса електронної пошти є коректною і «NO» в іншому випадку.

**Примітка:** Наявність символів @ та . недостатньо для коректності email адреси, проте їх відсутність гарантовано спричиняє неправильний email.

**Sample Input 1:**

aaaa@bbb.com

**Sample Output 1:**

YES

**Sample Input 2:**

aaaa@bbbcom

**Sample Output 2:**

NO

**Sample Input 3:**

qwerty.com

**Sample Output 3:**

NO

```python
email = input()

if "@" in email:
    if email.split("@")[1].count(".") > 0:  # Перевіряємо, чи ім'я домену містить хоча б одну крапку
        print("YES")
    else:
        print("NO")  # Якщо немає крапки у домені, то адреса не є коректною
else:
    print("NO")
```

**Чи відпочиваємо?**

Напишіть програму, яка зчитує один рядок, після чого виводить "YES", якщо у введеному рядку є підрядок "субота" або "неділя", і "NO" в іншому випадку.

**Формат вхідних даних**

**Формат вихідних даних**

**Sample Input 1:**

Який сьогодні день тижня?

Була субота, і йому хотілося якнайшвидше поїхати додому.

**Квадратне рівняння**

Дано три дійсних числа *a, b, c*. Напишіть програму, яка знаходить дійсні корені квадратного рівняння. Нагадаємо загальний вигляд квадратного рівняння:

*ax² + bx + c = 0*

Формула для розрахунку дискримінанта:

*D = b² - 4 * a * c*

Форумула для знаходження коренів:

*x1 = ( -b - sqrt(D) ) / 2a*

*x2 = ( -b + sqrt(D) ) / 2a*

**Формат вхідних даних**

На вхід програмі подається три дійсних числа, де *a не дорівнює 0*, кожне на окремому рядку.

**Формат вихідних даних**

Програма повинна вивести на екран дійсні корені рівняння.

**Примітка**. Якщо дискримінант мешне за нуль, програма повинна вивести рядок "Коренів немає".

**Sample Input 1:**

1

2

1

**Sample Output 1:**

-1.0

**Sample Input 2:**

1

-7.5

3

**Sample Output 2:**

0.4239663260874824

7.076033673912518

**Sample Input 3:**

4

-3

1

**Sample Output 3:**

Коренів немає

```python
a = float(input())
b = float(input())
c = float(input())

D = b ** 2 - 4 * a * c

if D > 0:
    x1 = (-b - D ** 0.5) / (2 * a)
    x2 = (-b + D ** 0.5) / (2 * a)
    print(x1)
    print(x2)
elif D == 0:
    x1 = -b / (2 * a)
    print(x1)
else:
    print("Коренів немає")
```
